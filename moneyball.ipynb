# Moneyball

Reading csv data.

```python
batting = pd.read_csv('data/Batting.csv', index_col='playerID', usecols=[0, 1, 3, 4, 6, 8, 15, 18, 20]).fillna(value=0)
```

```python
salaries = pd.read_csv('data/Salaries.csv', index_col='playerID', usecols=[3, 4])
```

```python
appearances = pd.read_csv('data/Appearances.csv', index_col='playerID', usecols=[0, 1, 3, 8, 9, 10, 11, 12, 13, 14, 15, 16])
```

Merge all csv files.

```python
b_a = appearances.merge(batting, left_index='playerID', right_index='playerID')
```

```python
baseball = b_a.merge(salaries, left_index='playerID', right_index='playerID')
```

```python
baseball.drop(baseball.columns[[11, 12]], axis=1, inplace=True)
```

Find the OBP.

```python
baseball['OBP'] = (baseball['BB'] + baseball['H'] + baseball['HBP'])/(baseball['AB'] + baseball['BB'] + baseball['HBP'] + baseball['SF'])
```

```python
baseball = baseball.sort_values(by='OBP',ascending=False)
```

```python
baseball[baseball['H'] + baseball['BB'] + baseball['HBP'] > 10].sort_values('OBP', ascending=False)
```

Find OBP by salary to find the players worth their salt.

```python
baseball['OBP/$'] = baseball['OBP'] / baseball['salary']
```

```python
baseball[baseball['H'] + baseball['BB'] + baseball['HBP'] > 10].sort_values('OBP/$', ascending=False)
```

Remove duplicates and players with perfect OBP. Replace infinite values.

```python
baseball = baseball.groupby(baseball.index).first()
baseball = baseball[baseball.OBP != 1]
baseball = baseball.replace(to_replace='inf', value=0)
```

Narrow down all players to team of 9 with a Pitcher, Catcher, Short Stop, First Base, Second Base, Third Base, Left Fielder, Center Fielder, and Right Fielder based on OBP by salary.

```python
def find_team(df):
    positions = ['G_p', 'G_c','G_1b', 'G_2b', 'G_3b', 'G_ss', 'G_lf', 'G_cf', 'G_rf']
    team = []
    for item in positions:
        team.append(df[df[item] > 50].sort_values(['OBP/$'], ascending=False)[:1])
    return team

team = find_team(baseball)
the_team = pd.concat(team)
the_team
```
